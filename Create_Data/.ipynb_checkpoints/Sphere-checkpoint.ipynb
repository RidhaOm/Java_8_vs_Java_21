<h1>Sphere</h1>

<h2>Bibliotheken importieren</h2>

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

<h2>Punktwolke erstellen</h2>

```python
def create_semi_sphere(num_vertices):
    # Erzeugt gleichmäßig verteilte Punkte auf der Oberfläche einer Halbkugel
    indices = np.arange(0, num_vertices, dtype=float) + 0.5
    phi = np.arccos(1 - 2*indices/num_vertices) / 2  # Teilen durch 2 begrenzt es auf eine Halbkugel
    theta = np.pi * (1 + 5**0.5) * indices

    x = np.sin(phi) * np.cos(theta)
    y = np.sin(phi) * np.sin(theta)
    z = np.cos(phi)

    # Filtere nur die Punkte, die auf der oberen Halbkugel sind
    upper_hemisphere = z >= 0
    return np.column_stack((x[upper_hemisphere], y[upper_hemisphere], z[upper_hemisphere]))

num_vertices = 642
vertices = create_semi_sphere(num_vertices)
```

<h2>Visualisierung der 3D-Punktwolke</h2>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2])#
ax.set_xlabel('X Koordinate')
ax.set_ylabel('Y Koordinate')
ax.set_zlabel('Z Koordinate')
plt.title('3D-Punktwolke für das Sphere-Modell')
plt.show()
```

<h2>Daten speichern</h2>

```python
np.savetxt("Sphere.csv", vertices, delimiter=",", fmt="%.18f")
```
