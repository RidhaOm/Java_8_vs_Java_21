<h1>Bunny</h1>

<h2>Bibliotheken importieren</h2>

```python
import open3d as o3d
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

<h2>Punktwolke aus ply-Datei erstellen</h2>

```python
# Laden des ply-Dateis
mesh = o3d.io.read_triangle_mesh("stanford_bunny.ply")

# Zugriff auf die Vertices und Konvertierung in ein NumPy-Array
vertices = np.asarray(mesh.vertices)
```

<h2>Visualisierung der 3D-Punktwolke</h2>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], c='r', marker='o')
ax.set_xlabel('X Koordinate')
ax.set_ylabel('Y Koordinate')
ax.set_zlabel('Z Koordinate')
plt.title('3D-Punktwolke des Bunny-Modells')
plt.show()
```

<h2>Daten speichern</h2>

```python
np.savetxt("Bunny.csv", vertices, delimiter=",")
```
