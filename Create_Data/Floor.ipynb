<h1>Floor</h1>

<h2>Bibliotheken importieren</h2>

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

<h2>Punktwolke erstellen</h2>

```python
def create_floor(num_vertices):
    # Erzeugt gleichmäßig verteilte Punkte auf einer flachen Oberfläche (Boden)
    # Erstelle ein Grid von Punkten in einem Quadrat
    side_length = int(np.sqrt(num_vertices))  # Seitenlänge für ein ungefähr quadratisches Layout
    x, y = np.meshgrid(np.linspace(-1, 1, side_length), np.linspace(-1, 1, side_length))
    z = np.zeros_like(x)  # Alle z-Koordinaten sind 0, da es sich um eine flache Oberfläche handelt

    # Flache das Grid zu einer Liste von Koordinaten ab
    floor_vertices = np.column_stack((x.flatten(), y.flatten(), z.flatten()))
    return floor_vertices

num_vertices = 10000
vertices = create_floor(num_vertices)
```

<h2>Visualisierung der 3D-Punktwolke</h2>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2])
ax.set_xlabel('X Koordinate')
ax.set_ylabel('Y Koordinate')
ax.set_zlabel('Z Koordinate')
plt.title('3D-Punktwolke für das Floor-Modell')
plt.show()
```

<h2>Daten speichern</h2>

```python
np.savetxt("Floor.csv", vertices, delimiter=",", fmt="%.18f")
```
