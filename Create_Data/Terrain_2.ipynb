<h2>Bibliotheken importieren</h2>

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

<h2>Punktwolke erstellen</h2>

```python
def create_terrain(num_vertices):
    # Erzeugt gleichmäßig verteilte Punkte auf einer unebenen Oberfläche (Terrain)
    side_length = int(np.sqrt(num_vertices))  # Berechne die Seitenlänge für ein quadratisches Layout
    x, y = np.meshgrid(np.linspace(-10, 10, side_length), np.linspace(-10, 10, side_length))
    
    # Generiere zufällige Höhenvariationen
    z = np.sin(x / 2) * np.cos(y / 2) + np.random.normal(0, 1, (side_length, side_length))
    
    # Flache das Grid zu einer Liste von Koordinaten ab
    terrain_vertices = np.column_stack((x.flatten(), y.flatten(), z.flatten()))
    return terrain_vertices

num_vertices = 263169
vertices = create_terrain(num_vertices)
```

<h2>Visualisierung der 3D-Punktwolke</h2>

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(vertices[:, 0], vertices[:, 1], vertices[:, 2], c=vertices[:, 2], cmap='viridis', s=1)
ax.set_xlabel('X Koordinate')
ax.set_ylabel('Y Koordinate')
ax.set_zlabel('Z Koordinate')
plt.title('3D-Punktwolke für das Terrain-Modell')
plt.show()
```

<h2>Daten speichern</h2>
